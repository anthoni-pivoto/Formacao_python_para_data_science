Filtro 1:
1. Apartamentos que possuem 1 quarto e aluguel menor que R$ 1200;


```python
df_apartamento['Quartos'] == 1
```

```python
selecao1 = df_apartamento['Quartos'] == 1
df_apartamento[selecao1]
```

```python
selecao2 = df_apartamento['Valor'] < 1200
df_apartamento[selecao2]
```

```python
selecao_final = (selecao1) & (selecao2)
df_apartamento[selecao_final]
```

```python
df_filtro1 = df_apartamento[selecao_final]
df_filtro1
```

Filtro 2:

2. Apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$ 3000 e área maior que 70 m².

```python
selecao_filtro2 = (df_apartamento['Quartos'] >= 2) & (df_apartamento['Valor'] < 3000) & (df_apartamento['Area'] > 70)
```

```python
df_filtro2 = df_apartamento[selecao_filtro2]
df_filtro2
```
